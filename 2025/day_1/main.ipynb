## Part 1

```python
import re

degrees = 50
total = 0
pattern = r"(\D+)(\d+)"

with open('input.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        match = re.match(pattern, line)
        if match:
            dir = match.group(1)
            amount = int(match.group(2))
        else:
            amount = 0
        print("Direction: ", dir)
        print("Amount: ", amount)
        if dir == 'R':
            degrees += int(amount)
        else:
            degrees -= int(amount)
        degrees %= 100
        if degrees == 0:
            total += 1
        
total
```

## Part 2

```python

degrees = 50
total = 0
pattern = r"(\D+)(\d+)"

with open('input.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        match = re.match(pattern, line.strip())
        if not match:
            continue
            
        direction = match.group(1)
        amount = int(match.group(2))
        
        prev_deg = degrees
        
        if direction == 'R':
            degrees += amount
            clicks = (degrees // 100) - (prev_deg // 100)
        else:
            degrees -= amount
            clicks = ((prev_deg - 1) // 100) - ((degrees - 1) // 100)

        total += clicks
        
        # Debug print to verify against your manual example
        # We perform % 100 only for printing the visual position, not for logic
        print(f"Move: {direction}{amount}")
        print(f"Pos: {prev_deg} -> {degrees} (Visual: {degrees % 100})")
        print(f"Clicks this turn: {clicks}")
        print(f"Total: {total}")
        print("-" * 20)

print(f"Final Password: {total}")
        
total
```
