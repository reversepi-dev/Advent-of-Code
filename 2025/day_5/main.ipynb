## Part 1

```python
with open("db.txt", "r") as file:
    rules = [[int(val) for val in line.strip().split('-')] for line in file.readlines()]

with open("available.txt", "r") as file_available:
    available = [int(line.strip()) for line in file_available.readlines()]

total_available = 0
for value in available:
    for rule in rules:
        if rule[0] <= value <= rule[1]:
            total_available += 1
            break
print("Total available: ", total_available)
```

## Part 2

```python
distinct_rules = []
rules.sort(key=lambda rl: (rl[0], rl[1]))
distinct_rules.append(rules[0])
for lb, ub in rules[1:]:
    if lb > distinct_rules[-1][1]:
        distinct_rules.append([lb, ub])
    else:
        distinct_rules[-1][1] = max(distinct_rules[-1][1], ub)

total_distinct_rules = 0
for rule in distinct_rules:
    total_distinct_rules += rule[1] - rule[0] + 1
print('Total distinct rules: ', total_distinct_rules)
```
