## Part 1

```python
with open('input.txt', 'r') as file:
    values = [list(digit) for digit in file.readlines()]

jolts = []
for value in values:
    highest_first_value, highest_first_index = None, None
    second_highest_value = None
    for index, digit in enumerate(value[:-2]):
        if highest_first_value is None or digit > highest_first_value:
            highest_first_value = digit
            highest_first_index = index
        if highest_first_value == '9':
            break
    for index, digit in enumerate(value[highest_first_index + 1:]):
        if second_highest_value is None or digit > second_highest_value:
            second_highest_value = digit

    jolts.append(highest_first_value + second_highest_value)
final_value = sum([int(jolt) for jolt in jolts])
print("Final jolts value: ", final_value)
```

## Part 2

Now we simply have to make it for instead of length 2 we have to make it for length 12. This will be done by changing the inner loop to go from 0 to length - str_digit_length + 1. Then we will simply inspect the string for any substrings of length str_digit_length. This can again be done in similar fashion but now we ensure that the length of the string can reach 12 for sure.

Approach:
1. Find the highest digit with which we can make a 12 digit number
2. Find the second highest digit to make the 12 digit number
3. This goes on until we reach length 12, and we always continue from the last found highest digit's index

```python
with open('input.txt', 'r') as file:
    values = [list(digit.strip()) for digit in file.readlines()]

jolts = []

for value in values:
    str_digit_length = 12
    current_index = 0
    result_digits = []

    for position in range(str_digit_length):
        highest_digit = None
        highest_digit_index = None

        search_end = len(value) - (str_digit_length - position) + 1

        for index in range(current_index, search_end):
            if highest_digit is None or value[index] > highest_digit:
                highest_digit = value[index]
                highest_digit_index = index
            if highest_digit == '9':
                break

        result_digits.append(highest_digit)
        current_index = highest_digit_index + 1

    jolts.append(''.join(result_digits))

final_value = sum([int(jolt) for jolt in jolts])
print("Final jolts value for length 12: ", final_value)
```
