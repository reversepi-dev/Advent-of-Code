## Part 1

These do not require a lot of explanation. Solutions speak for themselves.

```python
import numpy as np

with open("input.txt", "r") as file:
    data = np.array([line.strip().split() for line in file.readlines()])


def row_based_operation(row):
    if row[-1] == '+':
        return np.sum(row[:-1].astype(int))
    elif row[-1] == '*':
        return np.prod(row[:-1].astype(int))
    return 0

print("Answer:", int(np.sum(np.apply_along_axis(row_based_operation, 0, data))))
```

## Part 2

This one was a lot more complicated. The main difficult part was correctly parsing the input file including the spaces and when they are started. Therefore that is also the biggest chunk on which the code is made.

The final solution is also a bit different but that is mostly for doing a transpose operation on the existing matrix that was formed by the parsing and then applying the `join_row` function on each row.

```python
import numpy as np

def split_by_cracks(filename):
    with open(filename, 'r') as f:
        lines = [line.rstrip('\n') for line in f.readlines()]

    max_len = max(len(line) for line in lines)
    padded_lines = [line.ljust(max_len) for line in lines]
    num_rows = len(padded_lines)
    crack_indices = []

    for col in range(max_len):
        if all(padded_lines[row][col] == ' ' for row in range(num_rows)):
            crack_indices.append(col)

    blocks = []
    current_block_start = 0

    for col in range(max_len + 1):
        if col in crack_indices or col == max_len:
            if current_block_start < col:
                block = [list(line[current_block_start:col]) for line in padded_lines]
                if any(cell != "" for cell in block):
                    blocks.append(block)
            current_block_start = col + 1

    return blocks

subarrays = split_by_cracks('input.txt')

def join_row(row):
    return int("".join(row))

def row_based_operation_part_2(row):
    if "".join(row[-1]).strip() == '+':
        transposed_row = np.apply_along_axis(join_row, 1, row[:-1].transpose())
        return np.sum(transposed_row)
    elif "".join(row[-1]).strip() == '*':
        transposed_row = np.apply_along_axis(join_row, 1, row[:-1].transpose())
        return np.prod(transposed_row)
    return 0
total = 0
for subarray in subarrays:
    total += row_based_operation_part_2(np.array(subarray))
print("Answer:", total)
```
